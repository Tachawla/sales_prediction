```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

filename='/content/drive/MyDrive/sales_predictions.csv'
df=pd.read_csv(filename)
df.head()

#create histograms
fig1=plt.figure()
plt.hist(df['Item_Weight'],bins=[0,5,10,15,20,25], color='green',edgecolor='black')
plt.xlabel('Item Weight')
plt.ylabel('Frequency')

fig2=plt.figure()
plt.hist(df['Item_Visibility'],bins=10, color='blue',edgecolor='black')
plt.xlabel('Item Visibility')
plt.ylabel('Frequency')

fig3=plt.figure()
plt.hist(df['Item_MRP'],bins=10, color='red',edgecolor='black')
plt.xlabel('Item MRP')
plt.ylabel('Frequency')

fig4=plt.figure()
plt.hist(df['Item_Outlet_Sales'], bins=10, color='yellow', edgecolor='black')
plt.xlabel('Item_Outlet_Sales')
plt.ylabel('Frequency')

# boxplot
fig5=plt.figure()
df.boxplot(column='Item_Weight',color='green')

fig6=plt.figure()
df.boxplot(column='Item_Visibility',color='blue')

fig7=plt.figure()
df.boxplot(column='Item_MRP',color='red')

fig8=plt.figure()
df.boxplot(column='Item_Outlet_Sales',color='yellow')

fig9=plt.figure()
df.corr()
sns.heatmap(df.corr(),cmap='Purples')
```
